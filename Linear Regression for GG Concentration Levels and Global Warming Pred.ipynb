```python
import pandas as pd
import numpy as np
```

```python
# CO2 (carbon dioxide) is the Greenhouse Gas with highest atmospheric concentration level
# Units of CH4 concentration level: ppm (parts per million)
# Shown below is the CO2 data with only the relevant columns
co2 = pd.read_csv('global-co-concentration-ppm.csv')
co2[:10]
co2_17 = co2.loc[73:222, :]
co2_17 = co2_17.loc[:, ['Year', 'CO2 concentrations (NOAA, 2018)']]
co2_17
```

```python
# CH4 (methane) is the Greenhouse Gas with the second highest atmospheric concentration level
# Units of CH4 concentration level: ppb (parts per billion), not ppm. Conversion: 1 ppb = 0.001 ppm
# Shown below is the CH4 data with only the relevant columns
ch4 = pd.read_csv('ch4-concentration.csv')
ch4_17 = ch4.loc[:, ['Year', 'CH4 concentration (EEA & NOAA (2019))']]
ch4_17
```

```python
# N2O (nitrous oxide) is the Greenhouse Gas with the third highest atmospheric concentration level
# Units of N2O concentration level: ppb.
# Shown below is the N2O data with only the relevant columns
n2o = pd.read_csv('n2o-concentration.csv')
n2o_17 = n2o.loc[:, ['Year', 'N2O concentrations (annual average) (EEA, 2019)']]
n2o_17
```

```python
# Merged dataframe: all four independent variables (the year and the three greenhouse gas concentrations)
co2_ch4 = pd.merge(co2_17, ch4_17)
co2_ch4_n2o = pd.merge(co2_ch4, n2o_17)
co2_ch4_n2o
```

```python
# Global surface temperature data

#temps = pd.read_csv('GlobalTemperatures.csv')
#temps
```

```python
# Preprocessed data of annual average global surface temperatures; no missing values
temps = pd.read_excel("Temp est.xlsx")
temps
```

```python
# DATAFRAME fully merged, where all five features are present.
df = pd.merge(co2_ch4_n2o, temps)
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error,mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Split Data
df_fea = df.drop('Global Average Temperature', axis=1).values
df_temp = df['Global Average Temperature'].values

X_train,X_test,y_train,y_test = train_test_split(df_fea,
                                                 df_temp,
                                                 test_size=0.15,
                                                 random_state=42)

print(X_train.shape)
print(X_test.shape)

# normalize features
normalizer = StandardScaler()
X_train = normalizer.fit_transform(X_train)
X_test = normalizer.transform(X_test)
```

```python
# Linear REgression
# Train model
lr = LinearRegression()

lr.fit(X_train,y_train)

print("Bias is "+str(lr.intercept_))
print("Coefficients  are "+str(lr.coef_))

y_train_pred = lr.predict(X_train)

mae = mean_absolute_error(y_train_pred,y_train)
mse = mean_squared_error(y_train_pred,y_train)
rmse = np.sqrt(mse)

print('Prediction for training set:')
print('MAE is: {}'.format(mae))
print('MSE is: {}'.format(mse))
print('RMSE is: {}'.format(rmse))
```

```python
#4. evaluate the model
y_test_pred = lr.predict(X_test)

mae = mean_absolute_error(y_test_pred,y_test)
mse = mean_squared_error(y_test_pred,y_test)
rmse = np.sqrt(mse)

print('Prediction for testing set:')
print('MAE is: {}'.format(mae))
print('MSE is: {}'.format(mse))
print('RMSE is: {}'.format(rmse))


labels = ['Year 1', 'Year 2', 'Year 3', 'Year 4', 'Year 5']
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, y_test[0:5], width, label='ground truth')
rects2 = ax.bar(x + width/2, y_test_pred[0:5], width, label='prediction')

ax.set_ylabel('Temp')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, y_test[5:10], width, label='ground truth')
rects2 = ax.bar(x + width/2, y_test_pred[5:10], width, label='prediction')

ax.set_ylabel('Temp')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
